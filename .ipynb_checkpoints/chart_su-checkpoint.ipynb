```python
import pandas as pd
import matplotlib.pyplot as plt
import platform
import re
import seaborn as sns
# 한글 폰트 설정
if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic')
elif platform.system() == 'Darwin':
    plt.rc('font', family='AppleGothic')
else:
    plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 파일 경로
file_path = "data/raw_data/weather_accident.csv"

# CSV 불러오기 (정확한 헤더 줄을 반영)
df = pd.read_csv(file_path, encoding='utf-8', skiprows=2)
df.columns = df.columns.astype(str).str.strip()

# '항목' 컬럼 확인 및 필터링
if '항목' not in df.columns:
    raise ValueError(f"'항목' 컬럼이 존재하지 않습니다. 실제 컬럼들: {df.columns.tolist()}")

df = df[df['항목'].astype(str).str.strip() == '발생건수 (건)'].copy()

weather_types = ['맑음', '흐림', '비', '안개', '눈']
year_map = {}          # {접미 숫자:int → 연도:int}

# 1) 열 이름 스캔하며 “날씨.숫자” 패턴 찾기
pattern = re.compile(r'^(맑음|흐림|비|안개|눈)\.(\d+)$')
for col in df.columns:
    m = pattern.match(col)
    if m:
        weather, idx = m.group(1), int(m.group(2))
        year_map.setdefault(idx, []).append(col)

# 2) idx ↔ 연도 매핑 추정 (예: 가장 작은 idx → 2013)
min_idx = min(year_map)
idx_to_year = {idx: 2013 + (idx - min_idx) for idx in year_map}

# 3) 2013~2023에 대응하는 열만 수집
filtered_cols = []
for idx, year in idx_to_year.items():
    if 2013 <= year <= 2023:
        filtered_cols.extend(year_map[idx])

# 쉼표 제거 및 숫자 변환
for col in filtered_cols:
    df[col] = df[col].astype(str).str.replace(',', '').str.strip()
    df[col] = pd.to_numeric(df[col], errors='coerce')

# 기상상태별 합계 계산
weather_totals = {}
for weather in weather_types:
    cols = [col for col in filtered_cols if weather in col]
    weather_totals[weather] = df[cols].sum().sum()

# 연도별 합계 계산
year_totals = {}
for idx, year in idx_to_year.items():
    cols = [col for col in filtered_cols if str(idx) in col]
    year_totals[year] = df[cols].sum().sum()
print("기상상태별 교통사고 발생건수:", weather_totals)

# 시각화
plt.figure(figsize=(10, 6))
plt.bar(weather_totals.keys(), weather_totals.values(), color='skyblue')
plt.title("2013~2023년 기상상태별 교통사고 발생건수")
plt.ylabel("발생건수 (건)")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# 1) CSV 불러오기
df = pd.read_csv('data/first_processing_data/Traffic_accidents_multi_metric_tidy.csv', encoding='utf-8-sig')

# 2) 연도·값 전처리
df['year'] = pd.to_numeric(df['year'], errors='coerce').astype(int)
df['value'] = pd.to_numeric(df['value'], errors='coerce').fillna(0).astype(int)
df = df[(df['year'] >= 2005) & (df['year'] <= 2023)]

# 3) 지표 목록
metrics = df['metric'].unique()

# 4) 각 지표별 연도별 합계로 플롯
for metric in metrics:
    sub = df[df['metric'] == metric]
    yearly_sum = sub.groupby('year')['value'].sum().sort_index()
    
    plt.figure()
    plt.plot(yearly_sum.index, yearly_sum.values, marker='o')
    plt.title(f'2005–2023년 {metric} 추이')
    plt.xlabel('연도')
    plt.ylabel(metric)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
